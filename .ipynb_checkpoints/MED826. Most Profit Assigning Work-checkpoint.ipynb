```python
class Solution:
    def maxProfitAssignment(self, difficulty: List[int], profit: List[int], worker: List[int]) -> int:
        dic = []
        ans=0
        dic = [[i,0] for i in profit]

        for j in range(len(dic)):
            dic[j][1] = difficulty[j]
        ##ALL GOOD

        dic.sort(key = lambda x:x[0], reverse = True)

        # print(dic, ' Dic')
        worker.sort()
        profit.sort(reverse = True)
        # print(worker, ' Worker')
        ans = 0
        for i in worker:
            for j in range(len(profit)):
                # print(j)
                if i < dic[j][1]:
                    continue
                else:
                    ans += dic[j][0]
                    break
        return(ans)
```
