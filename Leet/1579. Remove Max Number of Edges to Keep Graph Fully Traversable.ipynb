```python
class Solution:
    def maxNumEdgesToRemove(self, n: int, edges: List[List[int]]) -> int:
        alice = []
        bob = []
        both = []
        checker = {i for i in range(1,n+1)}
        
        for i in edges:
            if i[0] == 3:
                both.append((i[1],i[2]))
            if i[0] == 1:
                alice.append((i[1],i[2]))
            if i[0] == 2:
                bob.append((i[1],i[2]))
        
       
        bob = list(set(both + bob)) 
        alice = list(set(both + alice))        
        
        bob.sort()
        alice.sort()
        
        for i in range(5):
            alice = self.convert(alice)
            bob = self.convert(bob)
            bob.sort()
            alice.sort() 
        

        if (1,n) not in alice or (1,n) not in bob:
            return -1
        temp = []
        for i in alice:
            temp.append(i[0])
            temp.append(i[1])
        alice = set(temp)

        temp = []
        for i in bob:
            temp.append(i[0])
            temp.append(i[1])
        bob = set(temp)
       
        for i in range(n//2):
            both.sort()
            both = self.convert(both) 
        both = set(both)   
        print(29, both,alice,bob)
        
        if bob != checker or alice != checker:
            return -1

        count = len(both)
        count += len(bob) - len(both)
        count += len(alice) - len(both)
        count = len(edges) - count
        return count + 2
        return 0

    # def poss(self,arr,checker):
    #     temp = [1]
    #     for i in set(arr):
    #         temp.append(i[1])
    #     if temp != checker:
    #         return False
    #     else:
    #         return True

    def convert(self, arr):
        for j in range(len(arr)-1):
            for k in range(j+1, len(arr)):
                if arr[j][1] == arr[k][0]:
                    arr.append((arr[j][0],arr[k][1]))
                    arr.pop(k)
        return arr
```
