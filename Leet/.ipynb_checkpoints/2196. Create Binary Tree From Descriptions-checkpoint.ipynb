```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def createBinaryTree(self, descriptions):
        root = 0
        child = []
        parent = []

        for i in descriptions:
            if i[1] not in child:
                child.append(i[1])
            if i[0] not in parent:
                parent.append(i[0])
        for i in descriptions:
            if i[0] not in child:
                root = i[0]
                break
        dic = {}
        for i in parent:
            dic.setdefault(i,[None,None])
        for i in descriptions:
            if i[2] == 0:
                dic[i[0]][1] = i[1]
            else:
                dic[i[0]][0] = i[1]
        head = TreeNode()
        head.val = root

        return self.maketree(head, dic, root)
        
    def maketree(self, head, dic, root):
        if head != None and head.val in dic:
            if dic[head.val][0] != None:
                head.left = TreeNode(dic[head.val][0])
            if dic[head.val][1] != None:
                head.right = TreeNode(dic[head.val][1])
            self.maketree(head.left, dic, root)
            self.maketree(head.right, dic, root)
        if head != None and head.val == root:
            return head
        
```
