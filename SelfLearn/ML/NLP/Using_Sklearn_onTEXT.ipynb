```python
import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import movie_reviews
import pandas as pd
```

```python
data = movie_reviews
data.categories()
```

```python
arr = []
for i in data.fileids():
    arr.append([data.words(i),data.categories(i)])
arr[0]
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
stopw = stopwords.words('english')
stopw += list(punctuation)
```

```python
#cleaned the stopwords and punctuation
for i in arr:
    temp = [j for j in i[0] if j not in stopw]
    i[0] = temp
    i[1] = i[1][0]
```

```python
##Lemmatizing
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemm = WordNetLemmatizer()
for i in arr:
    for j in i[0]:
        j = lemm.lemmatize(j)
```

```python
#Shuffle
```

```python
import random
random.shuffle(arr)
arr[0]
```

```python
#Getting it in appropriate format for CountVectorizer
```

```python
text_documents = [" ".join(i[0]) for i in arr]
categories = [i[1] for i in arr]
categories
```

```python
#Splitting and vectorizing
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000, ngram_range = (1,2))
xtr, xte, ytr, yte = train_test_split(text_documents, categories)
a = cv.fit_transform(xtr)
b = cv.transform(xte)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'rbf', C = 6) ##Hyperparameters after GSCV
svc.fit(a,ytr)
```

```python
svc.score(b, yte) ##Score after optimization
```
