```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('train.csv')
data
```

```python
data.drop(['Cabin','Name','Ticket', 'Embarked','SibSp','Parch'], axis = 1, inplace=True)
data
```

```python
j = []
for i in data['Sex']:
    if i == 'female':
        j.append(5)
    else:
        j.append(0)
j = np.array(j)
data['Sex_Fixed'] = j
data
```

```python
data.drop('Sex', axis =1 , inplace = True)
data
```

```python
y = data['Survived']
x = data
x.drop('Survived', axis= 1, inplace = True)
```

```python
x, y
```

```python
x['Age'] = x['Age'].fillna(np.mean(x['Age']))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
x_train
```

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
```

```python
y_logreg_pred = clf.predict(x_test)
clf.score(x_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clff = LinearRegression()
clff.fit(x_train, y_train)
```

```python
y_linreg_pred = clff.predict(x_test)
clff.score(x_test, y_test)
```
