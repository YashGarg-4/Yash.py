```python
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
data = datasets.load_iris()
x = data.data
y = data.target
```

```python
xtr,xte,ytr,yte = train_test_split(x,y)
```

```python
clf = svm.SVC()
clf.fit(xtr,ytr)
```

```python
clf.score(xte,yte)
```

```python
clff = GridSearchCV(clf, {
    'C' : [i for i in range(20)],
    'kernel' : ['rbf', 'linear', 'polynomial']
})
clff.fit(x,y)
clff.cv_results_
```

```python
import pandas as pd
df = pd.DataFrame(clff.cv_results_)
df.sort_values(by=['rank_test_score'], inplace = True)
df
```

```python
clf = svm.SVC(C=7, kernel = 'rbf')
clf.fit(xtr,ytr)
clf.score(xte,yte)
```
