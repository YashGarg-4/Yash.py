```python
# Number of ppl killed per country
```

```python
import csv
dic = {}
with open('year2017.csv') as obj:
    data = csv.DictReader(obj, skipinitialspace = True)

    for i in data:
        if i["Country"] not in dic:
            if i['Killed'] == '':
                dic.setdefault(i["Country"],0)
            else:
                dic.setdefault(i["Country"],float(i["Killed"]))
        if i['Killed'] == '':
            continue
        else:           
            dic[i['Country']] += float(i['Killed'])
    for i in dic:
        print(i, end='  -->  ')
        print(dic[i])
```
